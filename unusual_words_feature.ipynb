```
# Build function of counting word frenquency
def wordcounts (dataframe): 
    words = []
    for index, row in dataframe.iterrows():
        words.extend(row['total_token_for_id'])   #2855788 words in total - train ; 51385 for test
    word_counts = Counter(words)
    #least_500 = word_counts.most_common()[:-1000000-1:-1]    
    return word_counts
```

```
#word_counts = wordcounts (df_train)
word_counts = wordcounts (df_test)
#last_500_words = [i[0] for i in least_500]
```

```
frequency = [] #16.43 frequency for all words - train
least_words = []
for key, value in word_counts.items():
  if value < 2:
    least_words.append(key)
    frequency.append(value)
```

```
# basic tokenization
def token(doc):
  return [token.text for token in nlp(doc)]

# remove only stop words
def remove_stop (doc):
  return [token for token in nlp(doc) if not token.is_stop]

# remove punctuation
def remove_punct (doc):
  return [token for token in nlp(doc) if not token.is_punct]

# remove stop words and punctuation
def remove_stop_and_punct (doc):
  return [token for token in nlp(doc) if not (token.is_punct or token.is_stop)]

# lemmatization on origin data
def token_lemma(doc):
  return [token.lemma_ for token in nlp(doc)]

# lemmatization on cleaned data
def lemma_all(doc):
  return [token.lemma_ for token in nlp(doc) if not (token.is_punct or token.is_stop)]
```

```
df_test['sent1_no_stop_punct_lemma'] = df_test['sent1'].apply(lemma_all)
df_test['sent2_no_stop_punct_lemma'] = df_test['sent2'].apply(lemma_all)
```
